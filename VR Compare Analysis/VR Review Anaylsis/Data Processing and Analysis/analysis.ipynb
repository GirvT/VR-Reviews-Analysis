This data was collected by my scraping program and represents data scraped from google reviews.
Notable data attributes include:
- [Date] which is a relative date to 6/25/2024 where it becomes less granular as the review gets older since google reviews only show relative dates (i.e. 4 years ago, 2 months ago, 1 week ago, 2 days ago, etc.)
- [Sentiment] which are compound scores from the sentiment analysis model VADER included in nltk package 
- [Responce] which is the business responce if it exists
